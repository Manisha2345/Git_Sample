```python
query = f'''select   cast(tradingday as date)[Date],count(cast(tradingday as date))[Count]  from BondPricing.dbo.Czech_Republic 
where  cast(downloadeddate as date) <='2023-08-08'   group by cast(tradingday as date) order by cast(tradingday as date) desc '''
```

```python
FERACK2 = DatabaseManager('FERACK2','Bondpricing')
```

```python
result = FERACK2.fetchArray(query)
result
lst = []
for i in result:
    lst.append({'Date': i[0],'Count' : i[1]})
lst
```

```python
import pandas as pd
df = pd.DataFrame(lst)
df
```

```python
import plotly.express as px
```

```python
import plotly.express as px
import pandas as pd


fig = px.line(df, x="Date", y="Count", title="Year Wise Graph") 
fig.show()

df = df.sort_values(by="Date")
fig = px.line(df, x="Count", y="Count", title="Year Wise Graph(Line)") 
fig.show()
```

```python
import plotly.graph_objects as go

x = df['Date']
y = df['Count']

# Use textposition='auto' for direct text
fig = go.Figure(data=[go.Bar(
            x=x, y=y,
            text=y,
            textposition='auto',
        )])

fig.show()
```
